```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (classification_report, accuracy_score, precision_score,
                             recall_score, f1_score, confusion_matrix)
```

```python
# Load the dataset (replace 'data.csv' with your dataset path)
data = pd.read_csv('D:/Internships/Codsoft/task 4/creditcard.csv')
print(data.head())
```

```python
# Exploratory Data Analysis
print(data.describe())
print(data.shape)
print(data.isnull().values.any())

countclass = pd.value_counts(data['Class'], sort=True)
print(countclass.plot(kind= 'bar', rot = 0))
```

```python
# Divinding data into fraud and normal
fraud = data[data['Class']==1]
print(fraud)
print('Fraud Cases: {}'.format(len(data[data['Class']==1])))
normal = data[data['Class']==0]
print(normal)
print('Valid Transactions: {}'.format(len(data[data['Class']==0])))
```

```python
# Exploring the New datasets
print(fraud['Amount'].describe())
print(normal['Amount'].describe())
```

```python
# Checking the Correlation and plotting it
corr = data.corr()
plot = plt.figure(figsize= (14,8))
sns.heatmap(corr, vmax= 10, square=True)
plot.show()
```

```python
# Splitting the data set into training and testing sets
x = data.drop(['Class'], axis=1)
xData = x.values
y = data['Class']
yData = y.values
xTrain, xTest, yTrain, yTest = train_test_split(xData, yData, test_size=0.2, random_state=42)
```

```python
# Random Forset Classifier
rfc = RandomForestClassifier()
rfc.fit(xTrain,yTrain)
```

```python
# Prediction
Pred = rfc.predict(xTest)
```

```python
# Evaluation
n_outliers = len(fraud)
n_errors = (Pred != yTest).sum()
print('The Model used is Random Forest Classifier')

accuracy = accuracy_score(yTest,Pred)
print('Accuracy Score is {}'.format(accuracy))

precision = precision_score(yTest, Pred)
print('Precision is {}'.format(precision))

recall = recall_score(yTest, Pred)
print('Recall score is {}'.format(recall))

f1 = f1_score(yTest, Pred)
print('F1-Score is {}'.format(f1))
```

```python
# Confusion Matrix
label = ['normal', 'fraud']
conf = confusion_matrix(yTest, Pred)
plt.figure(figsize=(12,12))
sns.heatmap(conf, xticklabels=label, yticklabels=label,
            annot = True, fmt = "d");
plt.title("Confusion Matrix")
plt.ylabel('True Class')
plt.xlabel('Predicted Class')
plt.show()
```
